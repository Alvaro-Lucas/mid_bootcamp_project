```python
from config import user, password
from pymongo import MongoClient
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
```

```python
#Casos
url = "https://es.statista.com/estadisticas/1100641/regiones-afectadas-por-el-covid-19-segun-los-casos-confirmados-espana/"
res = requests.get(url)
html = BeautifulSoup(res.text)

graph = html.select("table.table.hidden td")
data_cases = [texto.get_text() for texto in graph]
data_cases = [ca.replace("C.","Comunidad").replace("Región de ","") for ca in data_cases]
ccaa_cases = data_cases[::2]
cases = data_cases[1::2]
cases = [case.replace(".","") for case in cases]

dict_cases = {}
for i in range(len(ccaa_cases)):
    dict_cases[ccaa_cases[i]] = cases[i]
dict_cases
```

```python
#Deaths
url = "https://es.statista.com/estadisticas/1104281/fallecidos-a-causa-de-covid-19-por-region-espana/"
res = requests.get(url)
html = BeautifulSoup(res.text)
graph = html.select("table.table.hidden td")

data_deaths = [texto.get_text() for texto in graph]
data_deaths = [ca.replace("C.","Comunidad").replace("Región de ","") for ca in data_deaths]
ccaa_deaths = data_deaths[::2]
deaths = data_deaths[1::2]
deaths = [deaths.replace(".","") for deaths in deaths]

suma = sum([int(death) for death in deaths])
print(suma)

dict_deaths = {}
for i in range(len(ccaa_deaths)):
    dict_deaths[ccaa_cases[i]] = deaths[i]
dict_deaths
```

```python
#Poblacion

url = "https://es.statista.com/estadisticas/472413/poblacion-de-espana-por-comunidad-autonoma/"
res = requests.get(url)
html = BeautifulSoup(res.text)
graph = html.select("table.table.hidden td")

data_pop = [texto.get_text() for texto in graph]
data_pop = [ca.replace("C.","Comunidad").replace("Castilla - La Mancha","Castilla-La Mancha").replace("Región de ","") for ca in data_pop]
ccaa_population = data_pop[::2]
population = data_pop[1::2]
population = [people.replace(".","") for people in population]

dict_population = {}
for i in range(len(ccaa_population)):
    dict_population[ccaa_population[i]] = population[i]
dict_population
```

```python
#Vacunados

url = "https://datosmacro.expansion.com/otros/coronavirus-vacuna/espana-comunidades-autonomas"
res = requests.get(url)
html = BeautifulSoup(res.text)
graph = html.select("table#tb1 thead th")
data_vac_column = [texto.get_text() for texto in graph]
data_vac_column[0] = "Community"
print(f"{data_vac_column = }")

list_dic = []
ccaa_vac = {}
graph = html.select("table#tb1 tbody tr")
data_vac_html = [texto for texto in graph]
vac = [vac.select("td") for vac in data_vac_html]

correct_names = ["Andalucía", "Aragón", "Asturias", "Cantabria", "Ceuta", "Castilla y León", "Castilla-La Mancha", "Canarias", "Cataluña",
                "Extremadura", "Galicia", "Islas Baleares", "Murcia", "Comunidad de Madrid", "Melilla", "Navarra", "País Vasco", "La Rioja",
                "Comunidad Valenciana"]

i = 0
for ccaa in vac:
    ccaa_vac = {}
    for pos in range(len(data_vac_column)):
        if pos == 0:
            ccaa_vac[data_vac_column[pos]] = correct_names[i]
            ccaa_vac["Lat"] = coordinates[correct_names[i]][0]
            ccaa_vac["Long"] = coordinates[correct_names[i]][1]
        else:
            try:
                ccaa_vac[data_vac_column[pos]] = float(ccaa[pos].get_text().replace("%","").replace(".","").replace(",","."))
            except:
                ccaa_vac[data_vac_column[pos]] = ccaa[pos].get_text().replace("%","")
    list_dic.append(ccaa_vac)
    i += 1
    
list_dic
```
